# **Control Flow**
* if
* for
* while

````python
# if
score = 82
if score >= 80:
    print("pass")
else:
    print("fail")
````

````python
## multiple grades รันจากบนลงล่าง
score = 50
if score >= 80:
    print("A")
elif score >= 70:
    print("B")
elif score >= 60:
    print("C")
else:
    print("D")
````

````python
## get input
username = input("What's your username: ")
age = int(input ("How old are you: "))
````

````python
print(age)
type(age)
````

````python
## multiple conditions
## night club
has_id = True
age_over_18 = False
vip_customer = False

if has_id and (age_over_18 or vip_customer):
    print("Welcome to the club.")
else:
    print("Please come next time.")
````

````python
## for & while
items = ["bag", "pen", "gum"]

for item in items:
    if item == "bag":
        print("My Bag")
    else:
        print(item)
````

````python
## ทำ uppercase ยาวเกิน
result = []

for item in items:
    result.append(item.upper())
print(result)
````

````python
## list comprehension (R ทำไม่ได้)
[item.upper() for item in items]
````

````python
## while
while True:
    user_input = input("Continue: Y/N? ")
    if user_input == "N":
        print("Bye!")
        break
    else:
        print("Run!")
````

````python
## while another e.g.
count = 0
while count < 5:
    print("test")
    count += 1
````

````python
## new function cube()
def cube(base, power):
    return base**power

result = cube(5, 4) ##เอาค่า returnไปเป็น variable ได้ 625
print(result)
````

````python
## function can return multiple values
## using tuple unpacking
def random_fun(base):
    return base**2, base**3, base**4
````

````python
x, y, z = random_fun(5)
print(x,y,z, x+y+z)
````
